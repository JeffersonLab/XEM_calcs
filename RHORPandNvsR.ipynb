```python
matplotlib.rcParams.update({'font.size': 22})
matplotlib.rcParams.update({'legend.framealpha': 1})
from matplotlib.ticker import StrMethodFormatter
from matplotlib.ticker import AutoMinorLocator
```

```python
hold = genfromtxt('Density_Data/h2.density.data')
```

```python
hold = numpy.delete(hold, (0), axis=0)
print (hold)
```

```python
sized = 10
plt.figure(figsize=(11, 9))
plt.gca().yaxis.set_minor_locator(AutoMinorLocator())

plot(hold[:,0],hold[:,2],'-bo', markersize=sized, label='RHORP')
xlabel('R')
ylabel('RHORP')
legend(loc="center", bbox_to_anchor=(.5,-.3))
title('R vs RHORP of H2')
```

```python
hold2 = genfromtxt('Density_Data/b11.density.data')
```

```python
hold2 = numpy.delete(hold2, (0), axis=0)
sized = 10
plt.figure(figsize=(11, 9))
plt.gca().yaxis.set_minor_locator(AutoMinorLocator())

plot(hold2[:,0],hold2[:,1],'-bo', markersize=sized, label='RHORP')
plot(hold2[:,0],hold2[:,3],'-r.', markersize=sized, label='RHORN')
xlabel('R')
ylabel('RHORP')
legend(loc="center", bbox_to_anchor=(.5,-.3))
title('R vs RHORP and RHORN of B11')
```
